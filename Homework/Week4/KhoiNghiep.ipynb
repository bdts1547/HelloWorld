#Khởi nghiệp
Sau thời gian làm việc tại nhà (work from home) do tình hình dịch bệnh, công ty X nhận thấy đây là cơ hội để thay đổi mô hình làm việc của công ty để hướng tới tính đơn giản, hiệu quả cao hơn trong công việc. Cụ thể, công ty hiện tại gồm n nhân viên. Có m quan hệ dạng ($a_i$, $b_i$) trong đó người bi nhận công việc từ người ai và ai chỉ đạo cho bi. Mỗi nhân viên khi nhận việc có thể phân công cho những nhân viên do mình chỉ đạo. Nếu công viêc từ a có thể tới được b, ta gọi a là cấp trên của b và b là cấp dưới của a.

Sau khi họp đại hội cổ ccông trực tuyến, công ty X quyết định thực hiện tái cấu trúc như sau:
*   Thứ nhất, chọn ra một giám đốc, người này phải là người không có cấp trên và công việc mà người này phân công có thể tới được mọi người trong công ty;
*   Những người còn lại, mỗi người sẽ chỉ còn nhận việc trực tiếp từ đúng 1 người. Để tránh xáo trộn trong hoạt động sau khi thực hiện thay đổi, với mỗi người, tập những người là cấp trên của người đó phải không thay đổi (so với trước khi thực hiện tái cấu trúc).

# **4.Đánh giá độ phức tạp thuật toán**
Độ phức tạp: O(V + E)

*   V: số đỉnh của đồ thị
*   E: số cạnh của đồ thị





# **5.Programming**


```python
import collections
import sys
from collections import defaultdict
sys.setrecursionlimit(5*10**5)


def addEdge_findRoot(n, m, graph):
    collection = set()
    checkG = defaultdict(list)
    for i in range(m):
        u , v = map(int, sys.stdin.readline().strip().split())
        graph[u].append(v)
        #print(graph)
        if v not in collection:
            collection.add(v)
    #print(collection)
    if len(collection) != n - 1:
       sys.stdout.write('No')
       exit()
    # find root
    for v in range(1, n + 1):
        if v not in collection:
            return v

# Dùng đệ quy
def dfs(u, graph, d, ans, deep, checked):
    deep[u] = d
    checked[u] = True
    if (u not in graph): return
    for v in graph[u]:
        if (deep[v] < d + 1):
            ans[v] = u
        if checked[v] == False:
            dfs(v, graph, d + 1, ans, deep, checked)

# Dùng vòng lặp
def dfs2(n, root, graph):
    to_collect = set()
    collected = set()
    to_collect.add(root)
    ans = [0] * (n + 1)
    ans[root] = -1
    deep = [0] * (n + 1)
    deep[root] = 1
    while(len(to_collect) > 0):
        u = to_collect.pop()
        collected.add(u)
        for v in graph[u]:
            if deep[v] < deep[u] + 1:
                ans[v] = u
            if v not in collected:
                to_collect.add(v)
    return ans[1:]


    

def main():
    graph = defaultdict(list)
    n, m = map(int, sys.stdin.readline().strip().split())
    root = addEdge_findRoot(n, m, graph)
    ans = dfs2(n, root, graph)
    string = ""
    for i in ans:
        if i != 0: 
            string += str(i) + " "
        else:
            sys.stdout.write('No')
            exit()

    sys.stdout.write('Yes' + '\n')
    sys.stdout.write(string)

main()
```
